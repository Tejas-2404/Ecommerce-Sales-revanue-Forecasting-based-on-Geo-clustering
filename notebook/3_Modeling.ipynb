```python
train_df = df[df['Date']<'2024-01-01']
test_df = df[df['Date']>='2024-01-01']
```

```python
train_df.sample(5)
```

```python
test_df.sample(5)
```

```python
features = ['Product Category', 'GeoCluster','Latitude','Longitude', 
            'Year','Month','Day', 'DayOfWeek', 'IsWeekend'
]
```

```python
target_volume = 'Sales Volume'
target_revenue = 'Sales Revenue'
```

```python
from sklearn.preprocessing import LabelEncoder 
for col in ['Product ID', 'Product Category']:
    encoder = LabelEncoder()
    train_df[col] = encoder.fit_transform(train_df[col])
    test_df[col] = encoder.transform(test_df[col])
```

```python
encoding_dict = {label: int(code) for label, code in zip(encoder.classes_, range(len(encoder.classes_)))}
```

```python
import json
with open('label_encoding.json', 'w') as f:
    json.dump(encoding_dict, f)
```

```python
X_train = train_df[features]
X_test = test_df[features]
```

```python
y_train_vol = train_df[target_volume]
y_test_vol = test_df[target_volume]
```

```python
y_train_rev = train_df[target_revenue]
y_test_rev = test_df[target_revenue]
```

```python
from sklearn.linear_model import LinearRegression
from xgboost import XGBRegressor
from sklearn.ensemble import RandomForestRegressor

models = {
    #'LightGBM': LGBMRegressor(random_state=42),
    'XGBoost': XGBRegressor(random_state=42, n_jobs=-1),
    'RandomForest': RandomForestRegressor(random_state=42, n_jobs=-1),
    'LinearRegression': LinearRegression()
}

def evaluate_model(name, y_true, y_pred):
    print(f"\n📊 {name}")
    print("MAE :", mean_absolute_error(y_true, y_pred))
    print("RMSE:", np.sqrt(mean_squared_error(y_true, y_pred)))
    print("R²  :", r2_score(y_true, y_pred))

result = {}
```

```python
for name, model in models.items():
    print(f"\n🚀 Training {name} for Sales Volume...")
    model.fit(X_train, y_train_vol)
    y_pred_vol = model.predict(X_test)
    evaluate_model(f"{name} - Volume", y_test_vol, y_pred_vol)

    print(f"🚀 Training {name} for Sales Revenue...")
    model.fit(X_train, y_train_rev)
    y_pred_rev = model.predict(X_test)
    evaluate_model(f"{name} - Revenue", y_test_rev, y_pred_rev)

```
